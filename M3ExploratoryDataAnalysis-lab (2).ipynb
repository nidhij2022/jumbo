### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


```python
print(
    '\033[4mAt the bigining\033[0m, There are \033[1m',
    df["ConvertedComp"].isna().sum(),
    '\033[0m  missing values in Column ConvertedComp.' 

)
print(
    "The Column ConvertedComp has \033[1m",
    len(df["ConvertedComp"]),
    "\033[0m entries.",
)

```

## Outliers


### Finding outliers


The median of ConvertedComp before removing outliers is:

```python
df1['ConvertedComp'].median()
```

Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
df1['ConvertedComp'].plot(kind='box', figsize=(20,10))
plt.title('Box plot of Salalry in USD')
plt.ylabel('Salary')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
q3, q1 = np.percentile(df1['ConvertedComp'], [75 ,25])
#Compute the q-th percentile of the data along the specified axis.
iqr_ConvertedComp=q3 - q1
print('the Inter Quartile Range for the column ConvertedComp is {}' .format(iqr_ConvertedComp))
```

Find out the upper and lower bounds.


```python
max_ConvertedComp,min_ConvertedComp=df1['ConvertedComp'].max(),df1['ConvertedComp'].min()
print('the upper bound equals ', max_ConvertedComp,'and lower bound equals ',min_ConvertedComp )
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
print([q1 - 1.5 *iqr_ConvertedComp, q3 + 1.5 * iqr_ConvertedComp])
outlier_condition=((df1['ConvertedComp'] < (q1 - 1.5 *iqr_ConvertedComp)) | (df1['ConvertedComp'] > (q3 + 1.5 * iqr_ConvertedComp)))
outlier_condition.sum()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
outlier_condition=((df['ConvertedComp'] < (q1 - 1.5 *iqr_ConvertedComp)) | (df['ConvertedComp'] > (q3 + 1.5 * iqr_ConvertedComp)))
df[outlier_condition] = np.nan
```

The median of ConvertedComp before removing outliers is:

```python
df['ConvertedComp'].median()
```

mean ConvertedComp after removing outliers equals:

```python
df['ConvertedComp'].mean()
```

```python
df['Age'].plot(kind='box', figsize=(20,10))
plt.title('Box plot of Age')
plt.ylabel('Age')
plt.show()
```

We see 0 outlier below Q1.
